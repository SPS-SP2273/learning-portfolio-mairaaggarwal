```python
import os
import glob
import shutil
```

# Exercise 1

## Task 1

```python
month_dict = {
    'jan': '01', 'feb': '02', 'mar': '03',
    'apr': '04', 'may': '05', 'jun': '06',
    'jul': '07', 'aug': '08', 'sep': '09',
    'oct': '10', 'nov': '11', 'dec': '12'
}

filename = 'oct-08_singapore.txt'
parts = filename.split('_')
month, date, city = parts[0].split('-')[0], parts[0].split('-')[1], parts[1].split('.')[0]
numerical_month = month_dict.get(month.lower(), month)
new_filename = f"{numerical_month}-{date}-{city}.txt"

print(new_filename)
```

## Task 2

```python
def rename_my_file(old_file_name):
    parts = old_file_name.split('_')
    date_parts = parts[0].split('-')
    
    month = month_dict.get(date_parts[0].lower(), date_parts[0])
    day = date_parts[1]
    city = parts[1].split('.')[0].title()

    new_file_name = f"{month}-{day}_{city}.txt"
    return new_file_name
```

## Task 3

```python
file_list = ['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
             'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
             'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
             'jan-01_tokyo.txt']

for old_file_name in file_list:
    new_file_name = rename_my_file(old_file_name)
    print(f"{old_file_name} ----> {new_file_name}")
```

## Task 4

```python
path = os.path.join('os-collaboration-exercise-data', '*.txt')
print(path)
```

```python
glob.glob(os.path.join('*.txt'), recursive=True)
```

```python
txt_files = glob.glob(os.path.join('*.txt'), recursive=True)

filenames=[]

for file in txt_files:
    filename = file.split(os.path.sep)[-1]
    filenames.append(filename)

print(filenames)
```

```python
for old_file_name in filenames:
    new_file_name = rename_my_file(old_file_name)
    os.rename(old_file_name, new_file_name)
    print(old_file_name, '--->', new_file_name)
```

## Task 7

```python
all_files = glob.glob(os.path.join('*.txt'), recursive=True)

for file in all_files:
    city_name = file.split('_')[1].split('.')[0].title()
    shutil.move(file, os.path.join(city_name, file))
```
